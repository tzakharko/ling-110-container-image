## <span id="notebooks">Introducing Jupyter notebooks</span>

What you are looking at now is a *Jupyter notebook*, which combines text and code in a single interactive document. In this course, we will use a combination of Jupyter Notebooks, plain python (*.py*) files, and Markdown (*.md*) files.

Because a Jupyter notebook changes as you interact with it, you should make sure to **keep saving it**, just like you do on your computer. You can **save the file** by clicking the *save* symbol at the top-left corner of the notebook, or by holding the *Control* key (on Windows) or the *Command* key (on Mac) and pressing *S*. The file will automatically save periodically, but if you save it manually then you can be sure you won't lose work in case your internet or computer stops unexpectedly.

### Cells

A Jupyter notebook is made up of *cells*, which are blocks of text or code. These are the sections that get a blue strip along the left side as you click around the notebook.

Any cell in a notebook can be **hidden** by clicking on its blue strip. Clicking the blue strip again will make the cell **reappear**.

Cells in a notebook can also be freely reordered. To **move a cell higher or lower**, click and hold on the white space between the blue strip and the start of the cell, and drag it up or down. A horizontal blue bar will show you where the cell will move when you release the mouse.

### Text cells

A *text cell* usually contains text, though it may also contain other non-interactive things like images. A text cell doesn't have any special appearance, other than the blue strip indicating that it is a cell. This cell is an example.

If you double-click on a text cell, it will **switch from display mode to editing mode**. You can make any edits you like. You might find this useful for annotating your notes before or during class.

**Switch back from editing mode to display mode** by clicking the button with the *play* symbol ⏵ at the top of the notebook, or by holding the *Shift* key and pressing *Enter* (on Windows) or *Return* (on Mac).

Text cells use something called *Markdown* for formatting. You won't need to know any for this course, but if you are interested, you can find a good guide [here](https://www.markdownguide.org/basic-syntax/).

### Code cells

A *code cell* contains code that you can write and run interactively in the notebook. A code cell has a gray background and shows square brackets outside the cell <span style="font-family: monospace">[ ]</span> on the left side, which may contain an asterisk <span style="font-family: monospace">*</span> or a number. An asterisk means that the code cell is currently running. A number means that the code cell has previously been run, and it tells you when that was: the first cell you ran after you opened the notebook will have number 1, the second cell you ran will have number 2, and so on. There is an example code cell below.

You can **edit a code cell** by clicking anywhere in it and typing.

You can **run a code cell** by clicking on it and clicking the button with the *play* symbol ⏵ at the top of the notebook, or holding *Shift* and pressing *Enter* / *Return*.

When you run a code cell, its output is displayed underneath. This includes any messages you have printed, as well as the result returned by running the last line of code in the cell.

**Warning:** when you run a code cell, it will have access to everything from code cells that you have previously run. This means:  
1. if your current cell refers to something from another cell, **you need to run that cell first**;  
2. if you run cells out of order, your results may depend on that order and thus **may not be reproducible**.  

Consequently, you should always arrange cells in the order you expect them to be run, and only run them in that order.

### Managing cells

If you want to **add a new cell**, click on the cell immediately above where you want it to be inserted, and click on the *+* button at the top of the notebook. The new cell will be a code cell by default; if you want it to be a text cell instead, click on the drop-down menu that says *Code* at the top of the notebook, and select the *Markdown* option.

If you need to **delete a cell**, click on the white space between the blue strip and the start of the cell, and press the *D* key twice. **If you delete a cell by mistake and want to undo it**, press the *Z* key.

**If your code is taking too long to run or starts an infinite loop**, or if you have run code out of order and you want to **restart with a clean slate**, click the button with the *reload* symbol ⟳ at the top of the notebook. You will see a popup message saying *Restart Kernel?*: click the red *Restart* button. 

> After restarting, you will no longer have access to outputs from code that you ran previously. It is helpful to clear these previous outputs so that you don't get confused. With your mouse anywhere inside the notebook, right-click (on Windows) or hold the *Ctrl* key and click (on Mac) to open the menu, then click the *Clear All Outputs* button.
> 
> You will also need to re-run your code after restarting.
